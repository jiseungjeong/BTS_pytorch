합성곱 신경망은 크게
- 합성곱 층(합성곱 연산->활성화함수(e.g. ReLU))
- 풀링 층(풀링 연산)

으로 구성

# 1. 합성곱 신경망의 대두

다층 퍼셉트론으로도 이미지 처리를 할 수 있지만, 그러면 1-d 데이터로 입력받으며 이미지 픽셀의 공간적 구조 정보를 효과적으로 받을 수 없음.
이를 위해 합성곱 신경망을 사용 

# 2. 채널(Channel)

이미지는 **(높이, 너비, 채널)** 이라는 3차원 텐서임. **(세로방향픽셀수, 가로방향픽셀수, 색성분)**. 흑백 이미지는 채널 수가 1. 각 픽셀은 0~255값.
채널은 때로 depth로도 불림.

# 3. 합성곱 연산 (Convolution operation)
이미지 특징을 추출하는 역할을 함. **커널, 혹은 필터라는 행렬(주로 3x3, 5x5)**로 전체 이미지를 겹치며 훑으며 값을 모두 곱해서 모두 더한 값으로 출력함.
입력으로부터 커널을 사용하여 합성곱 연산을 통해 나온 결과를 **특성 맵(feature map)** 이라고 함.
스트라이드 -> 보폭

# 4. 패딩(Padding)
합성곱 연산을 했을 때 출력 결과가 점점 작아지는 것을 방지하기 위해 겉에 0으로 두르는 것과 같이 패딩을 넣어 출력의 크기가 입력의 크기와 동일하게 유지되게 함. 주로 0으로 채우는 제로 패딩을 사용함. 스트라이드가 1일 때, 3x3 커널에서는 1폭짜리 제로패딩을 사용하고, 5x5 커널에서는 2폭짜리 제로패딩을 사용함.

# 5. 가중치와 편향
### 1. 합성곱 신경망의 가중치
공간적 구조를 고려하여 합성곱 연산을 진행하기에, 가중치 연결도 공간적 구조가 고려된 채로 진행된다. 
합성곱 연산을 통해 특성 맵을 업도, 활성화 함수를 지나는 연산을 하는 곳ㅇ르 합성공 층이라고 한다.(은닉층이다.)
### 2. 합성곱 신경망의 편향
편향을 추가할 수도 있는데, 이는 커널을 적용한 뒤에 더해지고, 하나의 값만이 있어, 커널이 적용된 결과의 모든 원소에 같은 값이 더해진다.


# 6. 특성 맵의 크기 계산 방법
입력의 크기와 커널의 크기, 스트라이드 값만 알면 특성맵의 크기를 계산할 수 있다.
패딩이 추가되었다면 패딩의 크기도 고려해야한다.

# 7. 다수의 채널을 가질 경우의 합성곱 연산(3차원 텐서의 합성곱 연산)
만약 다수의 채널을 가진 입력 데이터(컬러)를 가지고 합성곱 연산을 진행한다면 커널의 채널 수도 입력의 채널 수만큼 존재해야 한다.

**입력 데이터의 채널 수 = 커널의 채널 수**

합성곱 연산을 각 채널마다 수행함.

그리고 각 채널의 특성맵을 모두 더하여 최종 특성맵을 얻음.

주의) 1개의 커널 안에 3개의 채널이 있는 것임.


# 8. 3차원 텐서의 합성곱 연산
합성곱 연산에서 다수의 커널을 사용하면, 사용한 커널 수는 합성곱 연산의 결과로 나오는 특성맵의 채널수가 됨.

가중치는 커널의 원소임.

가중치 매개변수의 총 수: 커널 가로 x 커널 세로 x 입력 데이터 채널 수 x 특성 맵 채널 수

# 9. 풀링(Pooling)
풀링 층에서는 특성 맵을 다운 샘플링하여 특성 맵의 크기를 줄이는 풀링 연산이 이루어짐.

일반적으로 2가지 풀링이 있음.
- 최대 풀링 (max pooling)
- 평균 풀링 (average pooling)

풀링 연산에서도 합성곱 연산과 마찬가지로 커널과 스트라이드의 개념을 가짐.

맥스 풀링은 커널과 겹치는 영역 안에서 최대값을 추출하는 방식으로 다운 샘플링함. 평균 풀링은 평균값을 추출함. 풀링은 합성곱 연산과 다르게 학습해야 할 가중치가 없으며 연산 후에 채널 수가 변하지 않음.
